```python
import pandas as pd
import numpy as np
from scipy.stats import ks_2samp

baseline = pd.read_parquet("../data/demand_enriched_baseline.parquet")
baseline.head()
```

```python
print(baseline["time_bucket"].min())
print(baseline["time_bucket"].max())
```

```python
new_data = pd.read_parquet("../data/demand_enriched_week4.parquet")
new_data.head()
```

```python
print(new_data["time_bucket"].min())
print(new_data["time_bucket"].max())
```

```python
print(new_data.shape)
arr_1 = new_data["trip_count"].values[:2]
arr_2 = np.array([1, 2])
print(arr_1)
print(arr_2)
print(arr_1 == arr_2)
print((arr_1 == arr_2).sum())
print(2 == (arr_1 == arr_2).sum())
```

```python
print(baseline["time_bucket"].min())
print(baseline["time_bucket"].max())
new_cutoff = pd.Timestamp("2026-02-02")
feb_data = new_data[new_data["time_bucket"] >= new_cutoff]
print(feb_data["time_bucket"].min())
print(feb_data["time_bucket"].max())
ks_2_test_results = ks_2samp(baseline["trip_count"], feb_data["trip_count"])
print(ks_2_test_results)
print(ks_2_test_results.statistic)
print(ks_2_test_results.pvalue)
```
